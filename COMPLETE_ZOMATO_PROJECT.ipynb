```python
## QUESTION NUMBER =>> 2-1-2
import pandas as pd
import matplotlib.pyplot as plt
from statistics import mean
```

```python
data=pd.read_csv('zomato.csv',encoding='ISO-8859-1' )
```

```python
df=data.copy()
df=df[df['Country Code']==1]
df.reset_index(inplace=True,drop=True)
```

```python
rating=list(df['Aggregate rating'])
```

```python
lii=[]
def cuisines_len(s):
    lii.append(len(s.split(",")))
```

```python
df['Cuisines'].apply(cuisines_len)
df["number of cuisine"]=lii
print()
```

```python
plt.scatter(lii,rating)
plt.xlabel('No. of cuisine')
plt.ylabel('Rating')
plt.title("cuisine vs rating")
```

```python
d={}
for i in range(len(lii)):
    if d.get(lii[i],0)==0:
        d[lii[i]]=[]
        d[lii[i]].append(rating[i])
    else:
        d[lii[i]].append(rating[i])
        
for i in d:
    d[i]=mean(d[i]) 
    
d_2={}
for i in range(1,9):
    d_2[i]=len(list(df["number of cuisine"][df["number of cuisine"]==i]))
```

```python
x=[]
y=[]
z=[]
for i in range(1,9):
    x.append(i)
    y.append(d[i])
    z.append(d_2[i])
plt.scatter(x,y,s=180,c='y',alpha=0.5)
for i in range(0,8):
    plt.text(x[i]-0.2,y[i]-0.02,z[i])
plt.xlabel("No. of cuisine served")
plt.ylabel("AVG. rating")
plt.title("Cuisine served vs number of rating")
```

```python
d={}
for i in range(8652):
    city=df.loc[i,"City"]
    vote=df.loc[i,"Votes"]
    rating=df.loc[i,"Aggregate rating"]
    if vote>=0 and rating>=0:
        if d.get(city,0)==0:
            d[city]=[[],[]]
            d[city][0].append(vote)
            d[city][1].append(rating)
        else:
            d[city][0].append(vote)
            d[city][1].append(rating)        
```
